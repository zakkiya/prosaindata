# Piplines

**Import Database**

Dataset yang digunakan adalah dataset iris.
Tersdapat 4 kolom yaitu sepallength,sepalwidth, petallength, petalwidth.

Database yang digunakan adalah
1. PostgreSQL
2. Elephant
3. MySql
4. SQLserver

**Akses Database**

Akses database menggunakan Ngrok.
Di Ngrok disetting TCP port untuk masing-masing database.Kemudian di ngrok.yml ditambahkan sebuah tunnel untuk mewakili port database.


tunnels:
  mysql:
    addr: 3306
    proto: tcp
  postgre:
    addr: 5432
    proto: tcp

Kemudian untuk mengakses databases sesuai tcp yang muncul di ngrok (diganti host & port)

**Get Data from Database**

1. PostgreSQL

Untuk mengakses database PostgreSQL local menggunakan Ngrok. Kemudian menggunakan query untuk get kolom yang ada didatabase PostgreSQL. Kolom yang digunakan adalah sepallength. Library yang digunakan untuk mengakses database ini yaitu import psycopg2.

2. Elephant

Untuk mengakses database Elephant yang ada di postgreSQL, menggunakan library import psycopg2. Kemudian didalam library tersebut ada parameter yang harus dimasukkan. Kolom yang digunakan adalah sepalwidth.

3. MySQL

Untuk mengakses database MySQL menggunakan library mysql-connector. Supaya dapat diakses publik,  menggunakan Ngrok. Didalam parameter mysql.connector disetting host dan port sesuai ngrok yang disediakan. Data yang diambil adalah kolom petallength.

4. CSV

Untuk mengakses data csv menggunakan pandas. Kolom yang digunakan adalah petalwidth dan class.

Untuk menggabungkan dataframe menggunakan library pandas.
