```python
import string 
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from symspellpy import SymSpell ,Verbosity
import os
import requests
import numpy as np
```

```python
for i in data :
  data[data.index(i)] = i.split()
```

```python
#stopword
factory = StopWordRemoverFactory()
stopwords =  factory.create_stop_word_remover()
print(factory.get_stop_words())
```

```python
#prosesstopword
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = stopwords.remove(j)
```

```python
#lowercase
punctuation = "!@#$%^&*()/\[]=_+<>?:.,;" 
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",punctuation)).lower()
for i in data:
  for j in i:
    try:
      if(j[0] =='-'):
        data[data.index(i)][i.index(j)] = j.replace('-','')
    except:
      pass
#hapusnumber
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",string.digits))
#hapuskarakterkosong
for i in data:
  for j in i:
    try:
      if(len(j) <=2):
        data[data.index(i)][i.index(j)] = data[data.index(i)].remove(j)
    except:
      pass
for i in data:    
  for j in range(i.count('')):
    i.remove('')
  for j in range(i.count(None)):
    i.remove(None)
```

```python
for i in data:    
  for j in range(i.count('-')):
    i.remove('-')
    for j in range(i.count(None)):
      i.remove(None)
punctuation = "!@#$%^&*()â€™_+<>?:.,;!@#$%^&*()/\[]=_" 
for i in data:
  for j in i:
    data[data.index(i)][i.index(j)] = j.translate(str.maketrans("","",punctuation)).lower()
for i in range(len(data)):
  data[i] = " ".join(data[i])
```

```python
#stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()
for i in data:
  data[data.index(i)] = stemmer.stem(i)
```
